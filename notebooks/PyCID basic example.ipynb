```python
cid = pycid.CID([('A', 'P'), ('A', 'Ac'), ('P', 'Ac')],
          decisions=['P'],
          utilities=['Ac'])
```

```python
cid.draw()
```

```python
cid.add_cpds(A=pycid.discrete_uniform([0, 1]), # A uniform random CPD over its domain [0,1]
             Ac=lambda a, p: int(a == p), # specifies how 'U''s CPD functionally depends on its parents ('D' and 'S').
             P=[0, 1]) # specifies the domain of 'D' as [0,1].
```

```python
cid.draw()
```
